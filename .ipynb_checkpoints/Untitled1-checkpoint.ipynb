```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Initialize WebDriver with ChromeDriverManager
try:
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
except Exception as e:
    print(f"Error initializing WebDriver: {e}")
    exit()

def login_to_resy(driver):
    # Add code to log in if necessary
    pass

def book_table(driver, guests, date, time_slots):
    # Navigate to the Resy page for the restaurant
    driver.get("https://resy.com/cities/new-york-ny/venues/holywater?date=2024-07-19&seats=2")  # Replace with the actual URL

    try:
        # Wait for the page to load
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "div[class='date-picker']")))

        # Select the number of guests
        guests_dropdown = driver.find_element(By.CSS_SELECTOR, "select[aria-label='Guests']")
        guests_dropdown.click()
        guests_dropdown.send_keys(guests)
        guests_dropdown.send_keys(Keys.RETURN)

        # Select the date
        date_picker = driver.find_element(By.CSS_SELECTOR, "input[class='date-picker']")
        date_picker.clear()
        date_picker.send_keys(date)
        date_picker.send_keys(Keys.RETURN)

        # Select the time slot
        available_time_slots = driver.find_elements(By.CSS_SELECTOR, "button[class*='time-slot']")
        for slot in available_time_slots:
            for time_slot in time_slots:
                if time_slot in slot.text:
                    slot.click()
                    break

        # Wait for the reservation details page to load
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "div[class='reservation-summary']")))

        # Select the payment method
        payment_dropdown = driver.find_element(By.CSS_SELECTOR, "select[name='paymentMethod']")
        payment_dropdown.click()
        payment_dropdown.send_keys(Keys.DOWN)  # Adjust if you have multiple payment methods
        payment_dropdown.send_keys(Keys.RETURN)

        # Accept terms and conditions if required
        consent_toggle = driver.find_element(By.CSS_SELECTOR, "input[type='checkbox']")
        if not consent_toggle.is_selected():
            consent_toggle.click()

        # Click the "Reserve Now" button
        reserve_button = driver.find_element(By.CSS_SELECTOR, "button[class*='reserve-button']")
        reserve_button.click()

    except Exception as e:
        print(f"Error during booking process: {e}")

# Example usage
try:
    login_to_resy(driver)
    book_table(driver, "2 Guests", "07/19/2024", ["7:00 PM", "7:30 PM", "8:00 PM", "8:30 PM", "9:00 PM", "9:30 PM"])
    print("Booking completed successfully.")
except Exception as e:
    print(f"An error occurred: {e}")
finally:
    driver.quit()
```
